```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
def plot(url):
    df = pd.read_csv(url)
    df.d_rate = df.d_rate/100
    df.t = df.t - df.t[0]
    df.plot.line(x='t')
    
def pid_plot(url):
    df = pd.read_csv(url)
    df = df[['t','P', 'I', 'D', 'pid','speedpct']]
    df.t = df.t - df.t[0]
    df.plot.line(x='t')
    
def p_plot(url):
    df = pd.read_csv(url)
    df = df[['t','P','rate','speedpct']]
    df.t = df.t - df.t[0]
    df.plot.line(x='t')
```
